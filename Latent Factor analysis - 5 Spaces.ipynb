```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:\Python\Recommendation\Result\gamma=0.0007,lamda=0.01, steps=150\kmeans_qual5_analysis1.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
labels = ["S1","S2","S3","S4","S5"]
fs = 10  # fontsize
j = 9 # number of groups
nc = 3
k = 0
l = 0
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(80,120), sharey=True)
for i in range(0,j):
    a1 = df[df["Group"]==i][["S1","S2","S3","S4","S5"]].iloc[:,0:].values
    axes[k,l].boxplot(a1,labels=labels)
    axes[k,l].set_title('Group'+str(i), fontsize=fs)
    if i % nc == nc - 1:
        k = k + 1
        l = 0
    else:
        l = l + 1
        
   
```

```python
fig1, axes1 = plt.subplots(nrows=2, ncols=2, sharey=True)
df1 = df[df["Group"]==0][["S1","S2","S3","S4","S5"]]
sns.boxplot(df1,ax=axes1[0,0])
df1 = df[df["Group"]==1][["S1","S2","S3","S4","S5"]]
sns.boxplot(df1,ax=axes1[0,1])
```
